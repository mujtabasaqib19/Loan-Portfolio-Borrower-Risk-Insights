```python
import pandas as pd

# Load all datasets
loan_df = pd.read_csv("data/Loan.csv")
borrower_df = pd.read_csv("data/Borrower.csv")
loan_prod_df = pd.read_csv("data/Loan_Prod.csv")
borrower_prod_df = pd.read_csv("data/Borrower_Prod.csv")

# Summary EDA function
def summarize(df):
    return {
        "Shape": df.shape,
        "Columns": df.columns.tolist(),
        "Null Values": df.isnull().sum(),
        "Dtypes": df.dtypes,
        "Sample": df.head(3)
    }
```

```python
loan_summary = summarize(loan_df)
loan_summary
```

```python
borrower_summary = summarize(borrower_df)
borrower_summary
```

```python
loan_prod_summary = summarize(loan_prod_df)
loan_prod_summary
```

```python
borrower_prod_summary = summarize(borrower_prod_df)
borrower_prod_summary
```

```python
import numpy as np

loan_df = pd.read_csv("data/Loan.csv")

loan_df['isJointApplication'].fillna(0.0, inplace=True)
loan_df['loanAmount'].fillna(loan_df['loanAmount'].median(), inplace=True)

loan_df['term'] = loan_df['term'].astype(str)
loan_df['term'] = loan_df['term'].str.extract(r'(\d+)')
loan_df['term'] = loan_df['term'].astype(float)

missing_mask = loan_df['term'].isnull()
loan_df.loc[missing_mask, 'term'] = np.random.choice([30, 60], size=missing_mask.sum())
```

```python
loan_summary = summarize(loan_df)
loan_summary
```
